```
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#Chosen functions

f(x, y) = x^2 - x + y^2

```
def f(x, y):
  return x * (x - 1) + y**2
```

g(x, y) = y^3 + y^2 - x^2

```
def g(x, y):
  return y**2 * (y + 1) - x**2
```

h(x, y) = x^2 - y^2

```
def h(x, y):
  return x**2 - y**2
```

#Plotting f(x, y) - local min

```
# Graphs for f(x, y)

x = np.linspace(-3, 3, 15)
y = np.linspace(-3, 3, 15)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# 2D contour plot

plt.contourf(X, Y, Z, 50)
plt.figure()

# 3D contour plot

ax = plt.axes(projection='3d')
ax.grid(True)
ax.contour3D(X, Y, Z, 50)
```
