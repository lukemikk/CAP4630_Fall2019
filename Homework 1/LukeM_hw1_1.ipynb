#Problem 1

```
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

##Chosen functions

```
# f(x, y) = x^2 - x + y^2

def f(x, y):
  return x * (x - 1) + y**2


# g(x, y) = y^3 + y^2 - x^2

def g(x, y):
  return y**2 * (y + 1) - x**2


# h(x, y) = x^2 - y^2

def h(x, y):
  return x**2 - y**2
```

##Bonus: Partial derivatives of each function for gradients

```
# df/dx = 2x - 1

def df_dx(x):
  return 2 * x - 1

# df/dy = 2y

def df_dy(y):
  return 2 * y
```

```
# dg/dx = -2x

def dg_dx(x):
  return -2 * x

# dg/dy = 3y^2 + 2y

def dg_dy(y):
  return y * (3 * y + 2)
```

```
# dh/dx = 2x

def dh_dx(x):
  return 2 * x

# dh/dy = -2y

def dh_dy(y):
  return -2 * y
```

#Plotting f(x, y) - local min

```
# Graphs for f(x, y)

x = np.linspace(-3, 3, 15)
y = np.linspace(-3, 3, 15)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# 2D contour plot

plt.contourf(X, Y, Z, 50)
plt.figure()

# 3D contour plot

ax = plt.axes(projection='3d')
ax.grid(True)
ax.contour3D(X, Y, Z, 50)
```

##Plotting gradient of f(x, y)

```
X = np.arange(-3, 3, 2/5)
Y = np.arange(-3, 3, 2/5)
U, V = np.meshgrid(df_dx(X), df_dy(Y))

fig, ax = plt.subplots()
q = ax.quiver(X, Y, U, V)

plt.show()
```
